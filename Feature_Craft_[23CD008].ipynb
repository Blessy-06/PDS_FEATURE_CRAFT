```python
# 1. Discovering
# (i) list all column names that have trailing or leading spaces, check for inconsistent spellings or capitalization in categorical columns:
import pandas as pd

city_day_df = pd.read_csv('city_day.csv')
df = pd.read_csv('Crop_Production.csv')
print(df.head())
print(city_day_df.head())

cols_with_spaces = [col for col in df.columns if col.strip() != col]
print("\nColumns with extra spaces:", cols_with_spaces)

print("\nUnique District values:")
print(df['District'].unique())
```

```python
# (ii) Top five highest yiealding crops:
top_5_yield_crops = df.groupby('Crop')['Yield'].mean().sort_values(ascending=False).head(5)

print("Top 5 highest yielding crops (average yield):")
print(top_5_yield_crops)
```

```python
# (iii) Find all crop names ('Crop') that appear fewer than 300 times in the entire dataset:
crop_counts = df['Crop'].value_counts()

crops_less_than_300 = crop_counts[crop_counts < 300]

print("Crop names appearing fewer than 300 times:")
print(crops_less_than_300)
```

```python
# 2. Structuring
# (i) Rename columns by removing leading/trailing spaces:
df.columns = df.columns.str.strip()

print("Column names after removing spaces:")
print(df.columns)
```

```python
# (ii) Group by Crop and Season and calculate total Production and average Yield:
seasonal_performance = df.groupby(['Crop', 'Season']).agg(
    Total_Production=('Production', 'sum'),
    Average_Yield=('Yield', 'mean')
).reset_index()

print("Seasonal performance of each crop:")
display(seasonal_performance)
```

```python
# (iii) Calculate year-wise total production for each crop using pivot:
year_wise_production = df.pivot_table(
    index='Crop',
    columns='Crop_Year',
    values='Production',
    aggfunc='sum'
)

print("Year-wise total production for each crop:")
display(year_wise_production)
```

```python
# (iv) Calculate the highest average yield for each crop in each year:
average_yield_year_crop = df.groupby(['Crop', 'Crop_Year'])['Yield'].mean().reset_index()

highest_yield_crop_year = average_yield_year_crop.loc[average_yield_year_crop['Yield'].idxmax()]

print("Crop and Year with the highest average yield:")
print(highest_yield_crop_year)
```

```python
# 3. Cleaning
# (i) Change 'State' and 'District' to lowercase and remove leading/trailing spaces:
df['State'] = df['State'].str.lower().str.strip()
df['District'] = df['District'].str.lower().str.strip()

print("Unique values in 'State' after cleaning:")
print(df['State'].unique())

print("\nUnique values in 'District' after cleaning:")
print(df['District'].unique())
```

```python
# (ii) Calculate the average yield for each District and Crop combination:
district_crop_avg_yield = df.groupby(['District', 'Crop'])['Yield'].transform('mean')

df['Production'] = df['Production'].fillna(df['Area'] * district_crop_avg_yield)

print("Number of missing values in 'Production' after imputation:")
print(df['Production'].isnull().sum())
```

```python
# (iii) Filter the DataFrame for Wheat production in Odisha:
odisha_wheat_production = df[(df['Crop'] == 'Wheat') & (df['State'] == 'odisha')]

print("Wheat production in Odisha:")
display(odisha_wheat_production)
```

```python
# 4. Enrichment
# (i) Merge with External Data:
city_day_df_renamed = city_day_df.rename(columns={'City': 'State'})

city_day_df_renamed['State'] = city_day_df_renamed['State'].str.lower().str.strip()

merged_df = pd.merge(df, city_day_df_renamed, on='State', how='left')

print("Merged DataFrame:")
display(merged_df.head())
merged_df.info()
```

```python
# (ii) Calculate the median of 'Pollutant ratios':
median_pollutant_ratio = merged_df['Pollutant ratios'].median()

# Fill NaN values in 'Pollutant ratios' with the median
merged_df['Pollutant ratios'] = merged_df['Pollutant ratios'].fillna(median_pollutant_ratio)

print("Number of missing values in 'Pollutant ratios' after imputation:")
print(merged_df['Pollutant ratios'].isnull().sum())

print("\nDataFrame with NaN values in 'Pollutant ratios' filled with the median:")
display(merged_df[['SO2', 'NOx', 'Pollutant ratios']].head())
```

```python
# 5. Validating
# (i) Create a new column called Calculated_Yield:
merged_df['Calculated_Yield'] = merged_df['Production'] / merged_df['Area']

merged_df['Difference'] = merged_df['Calculated_Yield'] - merged_df['Yield']

sum_of_difference = merged_df['Difference'].sum()

print("Sum of the difference between Calculated_Yield and Yield:")
print(sum_of_difference)

print("\nDataFrame with 'Calculated_Yield' and 'Difference' columns:")
display(merged_df[['Production', 'Area', 'Yield', 'Calculated_Yield', 'Difference']].head())
```

```python
# (ii) Check for negative values in Area, Production, and Yield:
negative_values_check = merged_df[(merged_df['Area'] < 0) | (merged_df['Production'] < 0) | (merged_df['Yield'] < 0)]
print("Rows with negative Area, Production, or Yield values:")
display(negative_values_check)

production_area_check = merged_df[(merged_df['Production'] > 0) & (merged_df['Area'] == 0)]
print("\nRows where Production > 0 and Area == 0:")
display(production_area_check)

future_year_check = merged_df[merged_df['Crop_Year'] > 2024]
print("\nRows with Crop_Year in the future ( > 2024):")
display(future_year_check)
```
